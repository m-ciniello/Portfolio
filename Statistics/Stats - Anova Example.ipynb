# which <br />
The which() function will return the position of the elements(i.e., row number/column number/array index) in a logical vector which are TRUE. Unlike the other base R functions, the which() will accept only the arguments with typeof as logical while the others will give an error.  <br />

> x <- c(1,5,8,4,6) <br />
> x <br />
[1] 1 5 8 4 6 <br />
> which(x == 5) <br />
[1] 2 <br />
> which(x != 5) <br />
[1] 1 3 4 5 <br />


```r
for (i in 1:length(pos)) {
  these <- which(d$pos %in% pos[[i]])
  out   <- c(out, d[these,"OBP"])
  gp    <- c(gp, rep(POS[i], length(these)))
}
```

## Normality Test of residuals

mar   numerical vector indicating margin size c(bottom, left, top, right) in lines. default = c(5, 4, 4, 2) + 0.1 <br />
mai   numerical vector indicating margin size c(bottom, left, top, right) in inches <br />
pin   plot dimensions (width, height) in inches <br />

```r
par(mfrow=c(1,2))
par(pin = c(1.8,3))

qqnorm(mod$res,
       main = "Normality Test",
       xlab = "",
       ylab = "",
       pch = 10,
       cex = 0.7
      )
mtext("Theoretical Quantiles", 1, 2.2)
par(las = 0)
mtext("Residuals", 2, line = 3.3)
qqline(mod$res, col = 2)


hist(mod$res,
     main="Histogram of Residuals",
     xlab="Residuals",
     pch = 10,
     cex = 0.7)
lines(density(mod$res)) #Get a density curve to go along with your histogram
dev.off()

ad.test(mod$res)
```

## Normal probability plot of OBP for each field position

```r
par(mfrow=c(2,2)) 
qqnorm(out[gp == "OF"],
       main = "Outfielders",
       xlab = "",
       ylab = "",
       pch = 19,
       cex = 0.8,
       )
mtext("Theoretical Quantiles", 1, 2.2)
qqnorm(out[gp == "IF"],
       main = "In-fielders",
       xlab = "",
       ylab = "",
       pch = 19,
       cex = 0.8,
       )
mtext("Theoretical Quantiles", 1, 2.2)
qqnorm(out[gp == "DH"],
       main = "Designated Hitters",
       xlab = "",
       ylab = "",
       pch = 19,
       cex = 0.8,
       axes = FALSE)
axis(1)
axis(2, at = seq(0.3, 0.4, 0.05))
box()
mtext("Theoretical Quantiles", 1, 2.2)
qqnorm(out[gp == "C"],
       main = "Catchers",
       xlab = "",
       ylab = "",
       pch = 19,
       cex = 0.8,
       )
mtext("Theoretical Quantiles", 1, 2.2)
par(las = 0)
mtext("Residuals", 2, line = 3.3)
dev.off()
```
