```python
from scipy import stats
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

```

```python
#------------------------------------------------------------------------
#BINOMIAL
#------------------------------------------------------------------------
n=100
p=0.3
k = np.arange(0,100)
probs = stats.binom.pmf(k,n,p)

plt.plot(k,probs,'o-')
plt.title('BINOMIAL: n=%i, p=%.2f' % (n,p), fontsize=15)
plt.xlabel('Number of Successes')
plt.ylabel('Probability of Successes', fontsize = 15)
plt.show()

probs
```

```python
#------------------------------------------------------------------------
#BINOMIAL SIMULATION
#------------------------------------------------------------------------
binom_sim = stats.binom.rvs(n=10, p=0.3, size=10000)
print("Mean: %g" % np.mean(binom_sim))
print("SD: %g" % np.std(binom_sim, ddof=1))
plt.hist(binom_sim, bins=10,normed=True)
plt.xlabel("x")
plt.ylabel('density')
plt.show()
```

```python
#------------------------------------------------------------------------
#POISSON DISTRIBUTION
#------------------------------------------------------------------------

rate=1
n=np.arange(0,10)
probs=stats.poisson.pmf(n,rate)

plt.plot(n,probs,'o-')
plt.title('POISSON: $\lambda$ =%i' % rate)
plt.xlabel('Number of events')
plt.ylabel('Probability of Number of Events', fontsize = 15)
plt.show()
print(probs[0:3])
probs.sum()
```

```python
#------------------------------------------------------------------------
#NEGATIVE BINOMIAL DISTRIBUTION
#------------------------------------------------------------------------

k=range(0,10)
n=2
p=0.55
probs=stats.nbinom.pmf(k,n,p)


plt.plot(k,probs,'o-')
plt.title('Negative Binomial: $\lambda$ =%i' % rate)
plt.xlabel('Number of events')
plt.ylabel('Probability of Number of Events', fontsize = 15)
plt.show()

probs.mean()
```
