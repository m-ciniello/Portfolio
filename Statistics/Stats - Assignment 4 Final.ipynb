```python
import numpy as np
import pandas as pd
import scipy.stats as ss
import random
import matplotlib.pyplot as plt

# Simulate a population normally distributed with mean 100 and standard deviation 20 with a size of 1,000 
data_norm = ss.norm.rvs(100, 20.0, size=1000)
plt.hist(data_norm)
plt.show()
pop_mean = np.mean(data_norm)
print(pop_mean)
```

```python
# Take random 10 and 30 samples with no replacement of size 10, 30, and 100
# calculate the sample mean and standard deviation for each sample. 
# In total you will have 2x3 pairs (mean,std dev) of sample statistics. 
# Create a histogram of each sample mean distribution (6 in total)

#10 samples of (n=10, n=100, n=300)
norm_10_n10 = []
norm_10_n100 = []
norm_10_n300 = []
for i in range(10):
    norm_10_n10.append(np.mean(random.sample(list(data_norm),5)))
    norm_10_n100.append(np.mean(random.sample(list(data_norm),20)))
    norm_10_n300.append(np.mean(random.sample(list(data_norm),90)))

#30 samples of (n=10, n=100, n=300)
norm_30_n10 = []
norm_30_n100 = []
norm_30_n300 = []
for i in range(100):
    norm_30_n10.append(np.mean(random.sample(list(data_norm),5)))
    norm_30_n100.append(np.mean(random.sample(list(data_norm),20)))
    norm_30_n300.append(np.mean(random.sample(list(data_norm),90)))

#200 samples of (n=10, n=100, n=300)
norm_200_n10 = []
norm_200_n100 = []
norm_200_n300 = []
for i in range(1000):
    norm_200_n10.append(np.mean(random.sample(list(data_norm),5)))
    norm_200_n100.append(np.mean(random.sample(list(data_norm),20)))
    norm_200_n300.append(np.mean(random.sample(list(data_norm),90)))
    
fig = plt.figure(figsize=(10, 10))
ax1 = fig.add_subplot(3,3,1)
ax2 = fig.add_subplot(3,3,2)
ax3 = fig.add_subplot(3,3,3)
ax4 = fig.add_subplot(3,3,4)
ax5 = fig.add_subplot(3,3,5)
ax6 = fig.add_subplot(3,3,6)
ax7 = fig.add_subplot(3,3,7)
ax8 = fig.add_subplot(3,3,8)
ax9 = fig.add_subplot(3,3,9)

ax1.hist(norm_10_n10, histtype='bar', color=['green'])
ax2.hist(norm_30_n10, histtype='bar', color=['blue'])
ax3.hist(norm_200_n10, histtype='bar', color=['red'])
ax4.hist(norm_10_n100, histtype='bar', color=['green'])
ax5.hist(norm_30_n100, histtype='bar', color=['blue'])
ax6.hist(norm_200_n100, histtype='bar', color=['red'])
ax7.hist(norm_10_n300, histtype='bar', color=['green'])
ax8.hist(norm_30_n300, histtype='bar', color=['blue'])
ax9.hist(norm_200_n300, histtype='bar', color=['red'])

ax1.set_title('10 Samples')
ax2.set_title('100 Samples')
ax3.set_title('1000 Samples')
ax1.set_ylabel('n=5',rotation='horizontal')
ax4.set_ylabel('n=20',rotation='horizontal')
ax7.set_ylabel('n=90',rotation='horizontal')

plt.show()
```

```python
#create summary statistics

data_all = [norm_10_n10, 
            norm_10_n100, 
            norm_10_n300, 
            norm_30_n10, 
            norm_30_n100, 
            norm_30_n300, 
            norm_200_n10, 
            norm_200_n100, 
            norm_200_n300]

mean = [np.mean(x) for x in data_all]
minimum = [np.min(x) for x in data_all]
maximum = [np.max(x) for x in data_all]
diff_mean = [x-pop_mean for x in mean]
std_dev = [np.std(x) for x in data_all]
lengths = [len(x) for x in data_all]

data_all_dict = {'# of Samples':[10,10,10,100,100,100,1000,1000,1000],'n per sample':[5,20,90,5,20,90,5,20,90], 'mean':mean,
                 'Distance from pop mean':diff_mean,'Standard Deviation':std_dev,'Minimum':minimum,'Maximum':maximum}
data_all_df = pd.DataFrame(data_all_dict)

print('population mean: ', pop_mean)
data_all_df[['# of Samples','n per sample', 'mean','Distance from pop mean', 'Minimum','Maximum','Standard Deviation']]
```

```python
# Simulate normal, lognormal and uniform distributions with a size of 10,000

import numpy as np
import pandas as pd
import scipy.stats as ss
import matplotlib.pyplot as plt

# Simulate a population normally distributed with mean 100 and standard deviation 20 with a size of 1,000 
data_norm2 = ss.norm.rvs(10, 0.5, size=10000)
data_log = ss.lognorm.rvs(0.6,5,size=10000)
data_uni = ss.uniform.rvs(0.5, size=10000)

fig = plt.figure(figsize=(10, 5))
ax1 = fig.add_subplot(1,3,1)
ax2 = fig.add_subplot(1,3,2)
ax3 = fig.add_subplot(1,3,3)

ax1.hist(data_norm2, histtype='bar', color=['green'],bins=50)
ax2.hist(data_log, histtype='bar', color=['blue'],bins=50)
ax3.hist(data_uni, histtype='bar', color=['blue'],bins=50)

plt.show()
```
