```python
from datetime import datetime
import matplotlib.pyplot as plt
df = pd.read_excel('kpmg_data.xlsx')
df['date'] = df.apply(lambda x: datetime(x['Fyear'],x['Fperiod'],1), axis=1)

GIA = df.PCDesc == 'DA - Global Infrastructure'
df_gia = df[GIA].copy()

df_gia.sort_values('date', inplace=True)


df_gia.dropna(subset=['FeeRev_Performance_Gap'], axis=0, inplace=True)
```
