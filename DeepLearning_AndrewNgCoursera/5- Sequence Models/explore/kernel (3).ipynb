```python
# Load data
ngs = pd.read_csv("../input/NGS-2016-reg-wk7-12.csv")
roles = pd.read_csv("../input/play_player_role_data.csv")
```

## Merge and clean dataframes

```python
# Static vars
game =  175
play = 1335

# Filter and merge dataframes
ngs_cols = ['GSISID', 'Time', 'x', 'y', 'dis','o', 'dir', 'Event']
roles_cols = ['GSISID', 'Role']
ngs1 = ngs[(ngs.GameKey==game) & (ngs.PlayID==play)][ngs_cols]
roles1 = roles[(roles.GameKey==game) & (roles.PlayID==play)][roles_cols]

# Sort by 'Time' and 'GSISID'
df = pd.merge(left=ngs1, right=roles1, on='GSISID').sort_values(['GSISID','Time'])

# Add returner/role column as 'side'
returners = {'VRo','VRi','PDR1','PDR2','PDR3','PDL3','PDL1','PDL2', 'PLL','PLM', 'PLR','PR', 'VL'}
punters = {'PLW',   'PRT', 'GL', 'PRW', 'PLT',  'PLG', 'P','PC','PPR', 'PLS',  'PPL', 'PRG', 'GR'}
role_map = dict(list(zip(returners,['returning']*len(returners))) + list(zip(punters,['punting']*len(punters))))
df['side'] = df.Role.map(role_map)

# Create EventClean columns with ffill
df['EventClean'] = df.Event.fillna(method='ffill')

# Create 'action' column to isolate Events where game is in play
in_action = ['line_set', 'ball_snap', 'punt', 'punt_received','tackle']
df['action'] = df.EventClean.apply(lambda x: True if x in in_action else False)

df.head()
```
