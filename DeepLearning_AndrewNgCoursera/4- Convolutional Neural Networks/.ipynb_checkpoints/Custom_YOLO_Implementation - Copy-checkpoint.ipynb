# Working with a pretrained YOLO model

YOLO ("you only look once") is a popular algoritm because it achieves high accuracy while also being able to run in real-time. This algorithm "only looks once" at the image in the sense that it requires only one forward propagation pass through the network to make predictions. After non-max suppression, it then outputs recognized objects together with the bounding boxes.
In this exercise, you will learn how YOLO works, then apply it to car detection. Because the YOLO model is very computationally expensive to train, we will load pre-trained weights for you to use. 


Load the pretrained yolo model (`yolo.h5`) and do some exploration! Work with the keras package to investigate the inidivudal layers, weights, and overall model structure.

## Basic exploration

First things to know:
- The **input** is a batch of images of shape (m, 608, 608, 3)
- The **output** is a list of bounding boxes along with the recognized classes. Each bounding box is represented by 6 numbers $(p_c, b_x, b_y, b_h, b_w, c)$ as explained above. If you expand $c$ into an 80-dimensional vector, each bounding box is then represented by 85 numbers (80 binary class variables + 1 proba prediction + 4 bounding box numbers). 

We will use 5 anchor boxes. So you can think of the YOLO architecture as the following: IMAGE (m, 608, 608, 3) -> DEEP CNN -> ENCODING (m, 19, 19, 5, 85). **Remember, if the center/midpoint of an object falls into a grid cell, that grid cell is responsible for detecting that object.**

**For simplicity, we will flatten the last two last dimensions of the shape (19, 19, 5, 85) encoding. So the output of the Deep CNN is (19, 19, 425).**

```python
# First read in class and acnhors lists
def read_classes(classes_path):
    with open(classes_path) as f:
        class_names = f.readlines()
    class_names = [c.strip() for c in class_names]
    return class_names

def read_anchors(anchors_path):
    with open(anchors_path) as f:
        anchors = f.readline()
        anchors = [float(x) for x in anchors.split(',')]
        anchors = np.array(anchors).reshape(-1, 2)
    return anchors

classes_raw = read_classes("model_data/coco_classes.txt")
print("Classes:", len(classes_raw))

anchors_raw = read_anchors("model_data/yolo_anchors.txt")
print("Anchors:", anchors_raw.shape)
```

## `f1_yolo_head` func

Build `yolo_head` func, which takes in model output and converts it to...

```python
K.clear_session()
np.random.seed(1)
sess = K.get_session()
yolo = load_model("model_data/yolo.h5")
```

```python
# Build the function
def f1_yolo_head(feats, anchors, classes, norm_to_grid=True):

    # Extract number of anchors and classes
    num_anchors = len(anchors)
    num_classes = len(classes)
    
    # Create anchors varialbe and itinitalize it
    anchors_tf = K.variable(anchors, name='anchors_my_dude')
    K.get_session().run(tf.variables_initializer([anchors_tf]))
    
    # Reshape from (5,2) to (1,1,1,5,2) to match model output: (num_batches, height, width, num_anchors, box_params).
    anchors_tensor = K.reshape(anchors_tf, [1, 1, 1, num_anchors, 2])

    # Extract model dimensions of the image grid (19 X 19): assuming grid dimensions are in the middle (1,19,19,425)
    image_grid = K.shape(feats)[1:3] # In YOLO the height index is the inner most iteration.   
    # Reshape feats from (?,19,19,425) to (?,19,19,5,85)
    feats = K.reshape(feats, [-1, image_grid[0], image_grid[1], num_anchors, num_classes + 5])
    # Extract data from reshaped feats
    box_confidence = K.sigmoid(feats[..., 4:5]) # 4th elements of each 85 element vector
    box_xy = K.sigmoid(feats[..., :2]) # 1st and 2nd elements of each 85 element vector
    box_wh = K.exp(feats[..., 2:4]) # 3rd and 4th elements of each 85 element vector
    box_class_probs = K.softmax(feats[..., 5:]) # 6th to 85th elements of each 85 element vector
    
    # Dynamically adjust preditions to each spatial grid point and anchor size.
    if norm_to_grid:
        conv_height_index = K.arange(0, stop=image_grid[0])
        conv_height_index = K.tile(conv_height_index, [image_grid[1]])
        conv_width_index = K.arange(0, stop=image_grid[1])
        conv_width_index = K.tile(K.expand_dims(conv_width_index, 0), [image_grid[0], 1])
        conv_width_index = K.flatten(K.transpose(conv_width_index))
        conv_index = K.transpose(K.stack([conv_height_index, conv_width_index]))
        conv_index = K.reshape(conv_index, [1, image_grid[0], image_grid[1], 1, 2])
        conv_index = K.cast(conv_index, K.dtype(feats))    
        image_grid = K.cast(K.reshape(image_grid, [1, 1, 1, 1, 2]), K.dtype(feats))
        box_xy = (box_xy + conv_index) / image_grid
        box_wh = box_wh * anchors_tensor / image_grid
    
    return box_confidence, box_xy, box_wh, box_class_probs
```

```python
############ STEP 1: f1_yolo_head TENSORS INITIALIZATION ############
box_confidence, box_xy, box_wh, box_class_probs = f1_yolo_head(yolo.output, anchors_raw, classes_raw, norm_to_grid=True)

# print tensor shapes
for t in [box_confidence, box_xy, box_wh, box_class_probs]:
    print(t)
```

## `f2_yolo_boxes_to_corners` func
Build `yolo_boxes_to_corners` func, which takes `box_xy` predictions from `yolo_head` function and converts them to `((x1,y1),(x2,y2))` corner coordinate that can be plotted using PIL's `draw.rectangle` function.

```python
def f2_yolo_boxes_to_corners(box_xy, box_wh):
    """Convert YOLO box predictions to bounding box corners."""
    box_mins = box_xy - (box_wh / 2.)
    box_maxes = box_xy + (box_wh / 2.)

    return K.concatenate([
        box_mins[..., 1:2],  # y_min
        box_mins[..., 0:1],  # x_min
        box_maxes[..., 1:2],  # y_max
        box_maxes[..., 0:1]  # x_max
    ])
```

```python
############ STEP 2: f2_yolo_boxes_to_corners TENSORS INITIALIZATION ############
boxes = f2_yolo_boxes_to_corners(box_xy, box_wh)

# print tensor shapes
print(boxes)
```

## `yolo_filter_boxes` func
Build `yolo_filter_boxes` func, which takes the outputs of the `yolo_head` function (including the `box_xy` and `box_wh` tensors which are merged into corners by the `yolo_boxes_to_corners` func), and outputs the remaining boxes that have a **probability score above a specified threshold**.

```python
# Build the function
def yolo_filter_boxes(box_confidence, boxes, box_class_probs, threshold=.6):
    """Filters YOLO boxes by thresholding on object and class confidence.
    
    Arguments:
    box_confidence -- tensor of shape (19, 19, 5, 1)
    boxes -- tensor of shape (19, 19, 5, 4)
    box_class_probs -- tensor of shape (19, 19, 5, 80)
    threshold -- real value, if [ highest class probability score < threshold], then get rid of the corresponding box
    
    Returns:
    scores -- tensor of shape (None,), containing the class probability score for selected boxes
    boxes -- tensor of shape (None, 4), containing (b_x, b_y, b_h, b_w) coordinates of selected boxes
    classes -- tensor of shape (None,), containing the index of the class detected by the selected boxes
    
    Note: "None" is here because you don't know the exact number of selected boxes, as it depends on the threshold. 
    For example, the actual output size of scores would be (10,) if there are 10 boxes.
    """
    # Extract probabilities for each class (p * class_probas)
    box_scores = box_confidence * box_class_probs
    # Take argmax of grid box classes for each grid box (argmax returns element (class) in list where max occurs)
    box_classes = K.argmax(box_scores, axis=-1)
    # Take max proba scores for each grid box (Should be one for every grid box and every 
    box_class_scores = K.max(box_scores, axis=-1)
    
    filtering_mask = tf.greater(tf.to_float(box_class_scores), threshold) # float 

    # TODO: Expose tf.boolean_mask to Keras backend?
    boxes = tf.boolean_mask(boxes, filtering_mask)
    scores = tf.boolean_mask(box_class_scores, filtering_mask)
    classes = tf.boolean_mask(box_classes, filtering_mask)

    return boxes, scores, classes

#boxes, scores, classes = yolo_filter_boxes(box_confidence, boxes, box_class_probs, threshold=.6)
```

## `f4_scale_boxes`

```python
def scale_boxes(boxes, image_shape):
    """ Scales the predicted boxes in order to be drawable on the image"""
    height = image_shape[0]
    width = image_shape[1]
    image_dims = K.stack([height, width, height, width])
    image_dims = K.reshape(image_dims, [1, 4])
    boxes = boxes * image_dims
    return boxes
```

```python
# Load test images
root = 'images\\test_images'
images = [os.path.join(root, file) for file in os.listdir(root)]
images[120]
# Build the function
def preprocess_image(img_path, model_image_size=(608,608), normalize=255.0):
    # Get Image file type (we dont need this really, comment out for now)
    #image_type = imghdr.what(img_path)
    # Open image as PIL.Image file
    image = Image.open(img_path)
    # Resize to fit model inputs, using BICUBIC (3) as resample size 
    resized_image = image.resize(size=model_image_size, resample=Image.BICUBIC)
    # Convert to numpy array: shape should be (608,608,3)
    image_data = np.array(resized_image, dtype='float32')
    # Normalize (divide by max which should be 255)
    image_data /= normalize
    # Add batch dimension: shape should be (1, 608, 608, 3) 
    image_data = np.expand_dims(image_data, 0)  # Add batch dimension.
    # Return the raw image and the image numpy array!
    return image, image_data

img_raw, img_np = preprocess_image(images[120])
img_np.shape
```

## `generate_colors` func
Build the 'generate_colors' function, which generates a list of rgb tuples per class.

```python
# Experiment!
class_names = 'Mike Steve Evan Hailey'.split()
hsv_tuples = [(x / len(class_names), 1., 1.) for x in range(len(class_names))]
rgb_tuples = list(map(lambda x: colorsys.hsv_to_rgb(*x), hsv_tuples))
print(hsv_tuples)
print(rgb_tuples)
```

```python
# Build the function!
def generate_colors(class_names):
    # Create an hsv tuple (h,s,v) for each class
    hsv_tuples = [(x / len(class_names), 1., 1.) for x in range(len(class_names))]
    # Convert the hsv tuples into rgb tuples using map
    colors = list(map(lambda x: colorsys.hsv_to_rgb(*x), hsv_tuples))
    # Multiply each term by 255 (max rgb color)
    colors = list(map(lambda x: (int(x[0] * 255), int(x[1] * 255), int(x[2] * 255)), colors))
    np.random.seed(10101)  # Fixed seed for consistent colors across runs.
    np.random.shuffle(colors)  # Shuffle colors to decorrelate adjacent classes.
    np.random.seed(None)  # Reset seed to default.
    return colors

generate_colors(class_names)
```

## `draw_boxes` func
Build the `draw_boxes` function, which plots the predicted bounding boxes produced by the model.

```python
# Experiment!
test_img = images[120]
img_pil = Image.open(test_img)

# Create a PIL.Draw object: a simple 2D drawing interface for PIL images ('im' arg is the image you want to draw in).
draw = ImageDraw.Draw(im=img_pil)

# Load font OpenType font from 'font' file: PIL.ImageFont losad a TrueType or OpenType font from a file and creates a font object.
font = ImageFont.truetype(font='font/FiraMono-Medium.otf', size=20)
label = 'THIS IS A TEST WOOOT!!!'

# Use Draw object to draw some text and a rectangle
draw.rectangle([(300, 300),(600, 600)], fill=100)
draw.text(xy=[310,350],text=label, font=font)

# Use plt.imshow to play data (must convert to np.array first)
plt.figure(figsize=(15,15))
plt.imshow(np.array(img_pil))
```

```python
# Build the function
def draw_boxes(image, out_scores, out_boxes, out_classes, class_names, colors):
    # Extract font type from 'font' file folder, set size proportionate to Image ize
    font_size = np.floor(3e-2 * image.size[1] + 0.5).astype('int32')
    font = ImageFont.truetype(font='font/FiraMono-Medium.otf', size=font_size)
    # Set box thickness variables proportionate to image size
    thickness = (image.size[0] + image.size[1]) // 300

    for i, c in reversed(list(enumerate(out_classes))):
        # Extract class, box coordinates, and score from model outputs
        predicted_class = class_names[c]
        top, left, bottom, right = out_boxes[i]
        score = out_scores[i]
        label = '{} {:.2f}'.format(predicted_class, score)
        
        # Initialize ImageDraw.Draw object with image as input
        draw = ImageDraw.Draw(image)
        # draw.text size gets the size of a given string in pixels
        label_size = draw.textsize(label, font) 
        
        # Convert box coordinates to integers
        top = max(0, np.floor(top + 0.5).astype('int32'))
        left = max(0, np.floor(left + 0.5).astype('int32'))
        bottom = min(image.size[1], np.floor(bottom + 0.5).astype('int32'))
        right = min(image.size[0], np.floor(right + 0.5).astype('int32'))
        print(label, (left, top), (right, bottom))

        if top - label_size[1] >= 0:
            text_origin = np.array([left, top - label_size[1]])
        else:
            text_origin = np.array([left, top + 1])

        # My kingdom for a good redistributable image drawing library.
        for i in range(thickness):
            draw.rectangle([left + i, top + i, right - i, bottom - i], outline=colors[c])
        draw.rectangle([tuple(text_origin), tuple(text_origin + label_size)], fill=colors[c])
        draw.text(text_origin, label, fill=(0, 0, 0), font=font)
        del draw
```
